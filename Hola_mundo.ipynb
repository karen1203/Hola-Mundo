```python
#Hay que importar las librerías
import numpy  as np
from matplotlib import pyplot as plt
```

```python
#Definimos 2 arreglos de dimension
ys = 200+ np.random.randn(100)
x = np.arange(0,100,1)
```

```python
#Graficamos los rreglos en un plano 2D
plt.plot(x, ys, '-')
plt.fill_between(x,ys,195,where=(ys>195), facecolor='g', alpha=1)

#Agregamos titulo a la gráfica
plt.title("Sample Visualization")
plt.xlabel("eje de las x", size =16,)
plt.ylabel("eje de las y",size=16,)
plt.show()
```
